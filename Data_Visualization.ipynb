```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
C_df = pd.read_csv('/content/cn-C-array.csv',header=None)
C_df.head()
```

```python
G_df = pd.read_csv('/content/RNA_mean_part.csv')
G_df.head()
```

```python
G_data = G_df.iloc[:,1:].values
G_data[:,0]
```

```python
tags = G_df.iloc[:,0].values
Gi_tags = np.core.defchararray.add('Pre_G(i)_', tags.astype(str))
Gj_tags = np.core.defchararray.add('Post_G(j)_', tags.astype(str))
```

```python
C_data = {}
for i in range(len(Gi_tags)):
  C_data[Gi_tags[i]] = []
  C_data[Gj_tags[i]] = []
```

```python
C_DataFrame = pd.DataFrame(C_data)
C_DataFrame.head()
```

```python
len(list(C_data.keys()))
```

```python

# Sample C-matrix representing cell connections
C_matrix = C_df
# Sample G-matrix representing gene expression
G_matrix = G_data

# Initialize an empty list to store the extracted features and target
rows = []

# Iterate over the A-matrix to find cell connections and extract gene expression features
for i in range(len(C_matrix)):
    for j in range(len(C_matrix[i])):
        if i != j and C_matrix[i][j] > 0.1:
            # Extract gene expression features for connected cells
            features_i = pd.Series(G_matrix[:, i])
            features_j = pd.Series(G_matrix[:, j])
            # Add the features and connection status to the list
            new_row = {'connect': 1}
            for k, (feature_i, feature_j) in enumerate(zip(features_i, features_j), start=1):
                new_row[f'Pre_G(i)_{k}'] = feature_i
                new_row[f'Post_G(j)_{k}'] = feature_j
            # Append the new row to the list
            rows.append(new_row)
        elif i != j:
            features_i = pd.Series(G_matrix[:, i])
            features_j = pd.Series(G_matrix[:, j])
            # Add the features and connection status to the list for non-connected cells
            new_row = {'connect': 0}
            for k, (feature_i, feature_j) in enumerate(zip(features_i, features_j), start=1):
                new_row[f'Pre_G(i)_{k}'] = feature_i
                new_row[f'Post_G(j)_{k}'] = feature_j
            # Append the new row to the list
            rows.append(new_row)

# Create a DataFrame from the list of dictionaries
C_DataFrame = pd.DataFrame(rows)

# Print the DataFrame
print(C_DataFrame)
```

```python
C_DataFrame.columns = ['connect'] + list(C_data.keys())
```

```python
X = C_DataFrame.iloc[:,1:]
X
```

```python
y = C_DataFrame['connect']
```

```python
y
```

# **SPLITTING THE DATASET**

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

# **PRINCIPAL COMPONENT ANALYSIS**

```python

pca42 = PCA(n_components=42)
train_features_42 = pca42.fit_transform(X_train)
test_features_42 = pca42.transform(X_test)
```

# **VISUALIZATION OF PCA-COMPONENTS**

```python

if train_features_42.shape[1] >= 3:
    fig = plt.figure(figsize=(5,3),dpi=400)
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(train_features_42[:, 0], train_features_42[:, 1], train_features_42[:, 2])
    ax.set_xlabel('PC1')
    ax.set_ylabel('PC2')
    ax.set_zlabel('PC3')
    plt.show()
```

```python
X_pca = train_features_42
for i in range(min(3, X_pca.shape[1])):
    for j in range(i+1, min(3, X_pca.shape[1])):
        plt.scatter(X_pca[:, i], X_pca[:, j])
        plt.xlabel(f'PC{i+1}')
        plt.ylabel(f'PC{j+1}')
        plt.show()
```

```python
df_pca = pd.DataFrame(train_features_42, columns=[f'PC{i+1}' for i in range(train_features_42.shape[1])])
corr_matrix = df_pca.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

# **T-SNE DIMENSIONALITY REDUCTION**

```python
# Dimensionality Reduction with t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_pca)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1])
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.show()
```

# **K-MEANS CLUSTER ANALYSIS OF T-SNE COMPONENTS**

```python

# Cluster Analysis
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_pca)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=clusters, cmap='viridis')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.title('Clusters in t-SNE Space')
plt.show()
```
